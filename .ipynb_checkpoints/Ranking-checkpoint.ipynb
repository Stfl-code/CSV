```python
import streamlit as st
import gspread
import pandas as pd

# Connexion √† Google Sheets
gc = gspread.service_account_from_dict(st.secrets["gcp_service_account"])
sh = gc.open_by_key(st.secrets["sheet"]["id"])

# Onglet "R√©sultats"
resultats = sh.worksheet("resultats")

# Onglet "Joueurs"
joueurs = sh.worksheet("joueurs")
prenoms = joueurs.col_values(1)[1:]
noms = joueurs.col_values(2)[1:]
liste_joueurs = [f"{p} {n}" for p, n in zip(prenoms, noms)]

# Charger les r√©sultats existants
rows = resultats.get_all_records()
df = pd.DataFrame(rows)

# Onglets
tabs = st.tabs(["‚ûï Saisie des r√©sultats", "üìä Statistiques"])

# --- Onglet 1 : Saisie ---
with tabs[0]:
    st.header("Saisie d'un nouveau r√©sultat")

    with st.form("saisie_resultat"):
        joueur_A = st.selectbox("Joueur A", options=liste_joueurs)
        joueur_B = st.selectbox("Joueur B", options=[j for j in liste_joueurs if j != joueur_A])
        vainqueur = st.selectbox("Joueur A", options=[joueur_A, joueur_B])
        if vainqueur == joueur_A:
            score_A = 13
            st.number_input("Score A", min_value=13, max_value=13, value=13, disabled=True)
            score_B = st.number_input("Score B", min_value=0, max_value=12, value=0)
        else:
            score_B = 13
            st.number_input("Score B", min_value=13, max_value=13, value=13, disabled=True)
            score_A = st.number_input("Score A", min_value=0, max_value=12, value=0)
        submitted = st.form_submit_button("Enregistrer")

        if submitted:
           # V√©rifier qu'il n‚Äôexiste pas d√©j√† un r√©sultat entre ces deux joueurs
           existe_deja = False
           if not df.empty:
                existe_deja = (
                    ((df["joueur_A"] == joueur_A) & (df["joueur_B"] == joueur_B)).any()
                    or ((df["joueur_A"] == joueur_B) & (df["joueur_B"] == joueur_A)).any()
                )
                if existe_deja:
                    st.error("‚ö†Ô∏è Un r√©sultat existe d√©j√† pour cette paire de joueurs, s'il est n√©cessaire de le modifier veuillez contacter Stef-la-p√©tanque")
                else:
                    resultats.append_row([joueur_A, joueur_B, score_A, score_B, vainqueur])
                    st.success("‚úÖ R√©sultat enregistr√© avec succ√®s")

# --- Onglet 2 : R√©sultats r√©capitulatifs ---
with tabs[1]:
    st.header("Tableau des confrontations")

    # Cr√©ation d‚Äôune matrice vide
    recap = pd.DataFrame("", index=liste_joueurs, columns=liste_joueurs)

    for _, row in df.iterrows():
        A, B, sA, sB = row["joueur_A"], row["joueur_B"], row["score_A"], row["score_B"]
        recap.loc[A, B] = sA
        recap.loc[B, A] = sB

    st.dataframe(recap)

# --- Onglet 3 : Classement g√©n√©ral ---
with tabs[2]:
    st.header("Classement g√©n√©ral")

    stats = {j: {"Victoires": 0, "Points marqu√©s": 0, "Points encaiss√©s": 0} for j in liste_joueurs}

    for _, row in df.iterrows():
        A, B, sA, sB = row["joueur_A"], row["joueur_B"], row["score_A"], row["score_B"]

        # Mise √† jour des points
        stats[A]["Points marqu√©s"] += sA
        stats[A]["Points encaiss√©s"] += sB
        stats[B]["Points marqu√©s"] += sB
        stats[B]["Points encaiss√©s"] += sA

        # Attribution des victoires
        if sA > sB:
            stats[A]["Victoires"] += 1
        elif sB > sA:
            stats[B]["Victoires"] += 1

    # Conversion en DataFrame
    classement = pd.DataFrame(stats).T
    classement["Diff√©rence"] = classement["Points marqu√©s"] - classement["Points encaiss√©s"]

    # Tri : Victoires d'abord puis Diff√©rence
    classement = classement.sort_values(by=["Victoires", "Diff√©rence"], ascending=[False, False])

    st.dataframe(classement)
```
