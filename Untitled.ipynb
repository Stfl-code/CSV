```python
# Fonction pour calculer les stats en triplette
def calculer_stats_triplette(joueur_selectionne=None, partenaire1_selectionne=None, partenaire2_selectionne=None):
    stats = {j: {"Victoires": 0, "D√©faites": 0, "Points_marques": 0, "Points_encaisses": 0, "Diff": 0, "Toles_infligees": 0, "Toles_subies": 0} for j in liste_joueurs_complet}
    
    if not df_triplette.empty:  # df_triplette = r√©sultats de la feuille triplette
        for _, row in df_triplette.iterrows():
            vainq1 = row["vainqueur1"]
            vainq2 = row["vainqueur2"]
            vainq3 = row["vainqueur3"]
            adv1 = row["adversaire1"]
            adv2 = row["adversaire2"]
            adv3 = row["adversaire3"]
            score_v = row.get("score_vainqueur", 13)
            score_p = row.get("score_adversaire", 0)
            
            # √âquipes gagnante et perdante
            equipe_gagnante = [vainq1, vainq2, vainq3]
            equipe_perdante = [adv1, adv2, adv3]
            
            # Filtrer selon le joueur et partenaires s√©lectionn√©s
            if joueur_selectionne:
                if partenaire1_selectionne == "Tous" and partenaire2_selectionne == "Tous":
                    # Cas "Tous" : v√©rifier juste si le joueur a particip√©
                    if joueur_selectionne not in equipe_gagnante + equipe_perdante:
                        continue
                elif partenaire1_selectionne != "Tous" and partenaire2_selectionne != "Tous":
                    # Cas sp√©cifique : trio complet
                    trio = sorted([joueur_selectionne, partenaire1_selectionne, partenaire2_selectionne])
                    trio_gagnant = sorted(equipe_gagnante)
                    trio_perdant = sorted(equipe_perdante)
                    
                    if trio != trio_gagnant and trio != trio_perdant:
                        continue
                else:
                    # Un partenaire sp√©cifi√©, l'autre "Tous"
                    partenaire_specifie = partenaire1_selectionne if partenaire1_selectionne != "Tous" else partenaire2_selectionne
                    
                    # V√©rifier si le joueur et le partenaire sp√©cifi√© sont dans la m√™me √©quipe
                    joueur_et_partenaire_gagnants = joueur_selectionne in equipe_gagnante and partenaire_specifie in equipe_gagnante
                    joueur_et_partenaire_perdants = joueur_selectionne in equipe_perdante and partenaire_specifie in equipe_perdante
                    
                    if not (joueur_et_partenaire_gagnants or joueur_et_partenaire_perdants):
                        continue
            
            # Mettre √† jour les stats des vainqueurs
            for vainqueur in equipe_gagnante:
                if vainqueur in stats:
                    stats[vainqueur]["Victoires"] += 1
                    stats[vainqueur]["Points_marques"] += score_v
                    stats[vainqueur]["Points_encaisses"] += score_p
                    
                    if score_v == 13 and score_p == 0:
                        stats[vainqueur]["Toles_infligees"] += 1
            
            # Mettre √† jour les stats des perdants
            for perdant in equipe_perdante:
                if perdant in stats:
                    stats[perdant]["D√©faites"] += 1
                    stats[perdant]["Points_marques"] += score_p
                    stats[perdant]["Points_encaisses"] += score_v
                    
                    if score_v == 13 and score_p == 0:
                        stats[perdant]["Toles_subies"] += 1
    
    # Calculer la diff√©rence de points
    for j in stats:
        stats[j]["Diff"] = stats[j]["Points_marques"] - stats[j]["Points_encaisses"]
    
    return stats
```

```python
import networkx as nx
import itertools

def generer_appariements_suisse(nb_matchs, joueurs_liste):
    stats_tournoi = calculer_stats_tournoi()
    
    # Identifier les paires d√©j√† jou√©es ou programm√©es
    paires_existantes = set()
    if not tournoi_tat_df.empty:
        for _, row in tournoi_tat_df.iterrows():
            j1, j2 = sorted([row["joueur_1"], row["joueur_2"]])
            paires_existantes.add((j1, j2))
    
    # Cr√©er le classement
    classement = sorted(stats_tournoi.items(), key=lambda x: (x[1]["Victoires"], x[1]["Diff"]), reverse=True)
    
    # Cr√©er un dictionnaire de positions pour chaque joueur
    position_classement = {joueur: idx for idx, (joueur, _) in enumerate(classement) if joueur in joueurs_liste}
    
    # Cr√©er un graphe
    G = nx.Graph()
    G.add_nodes_from(joueurs_liste)
    
        # Ajouter des ar√™tes avec des poids bas√©s sur la proximit√© au classement
        for i, j in itertools.combinations(joueurs_liste, 2):
            paire = tuple(sorted([i, j]))
            
            # Ne pas cr√©er d'ar√™te si la paire existe d√©j√†
            if paire not in paires_existantes:
                # Calculer le poids : plus les joueurs sont proches au classement, plus le poids est √©lev√©
                pos_i = position_classement.get(i, len(joueurs_liste))
                pos_j = position_classement.get(j, len(joueurs_liste))
                
                # Distance au classement (invers√©e pour que proche = poids √©lev√©)
                distance = abs(pos_i - pos_j)
                poids = 1000 - distance  # Plus la distance est petite, plus le poids est √©lev√©
                
                G.add_edge(i, j, weight=poids)
    
        # Trouver le matching maximum avec poids maximum
        # maxcardinality=True garantit le nombre maximum de matchs
        matching = nx.max_weight_matching(G, maxcardinality=True)
    
    # D√©terminer le num√©ro du tour
    if tournoi_tat_df.empty:
        num_tour = 1
    else:
        # Compter les tours existants
        tours_existants = tournoi_tat_df["tour"].unique() if "tour" in tournoi_tat_df.columns else []
        if len(tours_existants) > 0:
            dernier_tour = max([int(t.replace("Tour ", "")) for t in tours_existants if isinstance(t, str) and t.startswith("Tour")])
            num_tour = dernier_tour + 1
        else:
            num_tour = 1
    
    # Convertir en liste de matchs
    nouveaux_matchs = [[i, j, f"Tour {num_tour}", "√† jouer", "", ""] for i, j in matching]
    
    # Limiter au nombre demand√© (normalement d√©j√† maximal)
    nouveaux_matchs = nouveaux_matchs[:nb_matchs]
    
    return nouveaux_matchs
```
